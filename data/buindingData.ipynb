```python
import os
import json
from dateutil import parser
import copy
import pickle
from datetime import datetime
import math
import time
from nltk import tokenize
import io
import pandas as pd
import en_core_web_sm
import random
from nltk.stem import WordNetLemmatizer
import nltk
from nltk.corpus import wordnet
import string
import csv

# Generating data for HAN (for building classifier to predict 17 crime tags)

random.seed(777777)

crimes = ["Oil Theft",
            "Microtrafficking",
            "Human Trafficking",
            "Human Smuggling",
            "Arms Trafficking",
            "Homicides",
            "Cyber Crime",
            "Criminal Migration",
            "Counterfeit",
            "Contraband",
            "Extortion",
            "Gender and Organized Crime",
            "Money Laundering",
            "European Organized Crime",
            "Illegal Mining",
            "Eco-Trafficking",
            "Kidnapping"]

crimesEncoding = {"Oil Theft":1,
            "Microtrafficking":2,
            "Human Trafficking":3,
            "Human Smuggling":4,
            "Arms Trafficking":5,
            "Homicides":6,
            "Cyber Crime":7,
            "Criminal Migration":8,
            "Counterfeit":9,
            "Contraband":10,
            "Extortion":11,
            "Gender and Organized Crime":12,
            "Money Laundering":13,
            "European Organized Crime":14,
            "Illegal Mining":15,
            "Eco-Trafficking":16,
            "Kidnapping":17}


data_file = 'crime_tags.json'
prevText = None

with open(data_file) as f:

    for line in f:
        d = json.loads(line)

        text = d['content']
        tags = d['tags']

        # Skip duplicated articles
        if text[0:50] == prevText:
            pass
        else:
            prevText = text[0:50] 

        label = None
        labelTag = None
        for tag in tags:
            if tag in crimes:
                label = crimesEncoding[tag]
                labelTag = tag
                break

        if label != None:

            rand = random.random()

            if rand <= 0.6666:

                with open('train.csv', mode='a') as employee_file:
                    employee_writer = csv.writer(employee_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
                    employee_writer.writerow([label, text])

            else:
                with open('test.csv', mode='a') as employee_file:
                    employee_writer = csv.writer(employee_file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
                    employee_writer.writerow([label, text])
```

```python
import os
import json
from dateutil import parser
import copy
import pickle
from datetime import datetime
import math
import time
from nltk import tokenize
import io
import pandas as pd
import en_core_web_sm
import random
from nltk.stem import WordNetLemmatizer
import nltk
from nltk.corpus import wordnet
import string
import csv

totalTrain = 0
totalTest = 0

labelsFreq = {}

crimesEncoding = {1:"Oil Theft",
            2:"Microtrafficking",
            3:"Human Trafficking",
            4:"Human Smuggling",
            5:"Arms Trafficking",
            6:"Homicides",
            7:"Cyber Crime",
            8:"Criminal Migration",
            9:"Counterfeit",
            10:"Contraband",
            11:"Extortion",
            12:"Gender and Organized Crime",
            13:"Money Laundering",
            14:"European Organized Crime",
            15:"Illegal Mining",
            16:"Eco-Trafficking",
            17:"Kidnapping"}


with open('train.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for row in csv_reader:
        label = row[0]
        totalTrain += 1
        if label in labelsFreq:
            labelsFreq[label] = labelsFreq[label] + 1 
        else: 
            labelsFreq[label] = 1 
        
with open('test.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    for row in csv_reader:
        
        label = row[0]
        totalTest += 1
        if label in labelsFreq:
            labelsFreq[label] = labelsFreq[label] + 1 
        else: 
            labelsFreq[label] = 1 
            
            
print('Total Training Sentences: ', totalTrain)
print('Total Testing Sentences: ', totalTest)
for k, v in labelsFreq.items():
    print(crimesEncoding[int(k)], ";", v)
```
